# SVR

```python
from sklearn.svm import LinearSVR
from sklearn.svm import SVR
import numpy as np 
import matplotlib.pyplot as plt
```

## Linear Reg

```python
# generate some training data
x_train=np.linspace(-1,1,101)
y_train=2.3*x_train+np.random.randn(*x_train.shape)*0.3 # * is for unpacking

# plot training data
plt.scatter(x_train,y_train,label="training data")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
svm_reg = LinearSVR(epsilon=0.5)
svm_reg.fit(x_train.reshape(-1,1),y_train)
```

```python
y_pred = svm_reg.predict(x_train.reshape(-1,1))
# plot training data
plt.scatter(x_train,y_train,label="training data")
plt.scatter(x_train,y_pred,label="model")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

## Polynomial Reg

```python
# generate some training data
x_train=np.linspace(-1,1,101)
num_coeff = 8
learning_rate = 0.01
training_epochs = 100
y_train_coeff = np.array([1, 3, 2, 6, 9, 8, 4, 5])

y_kernel = []
for i in range(num_coeff):
    term = np.power(x_train,i)
    y_kernel.append(term)
y_train = np.dot(y_train_coeff, y_kernel) + np.random.randn(*x_train.shape)*2    

# plot training data
plt.scatter(x_train,y_train,label="training data")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
svm_poly_reg = SVR(kernel="poly", degree=2, C=100, epsilon=0.5,gamma='auto',coef0=1.5)
svm_poly_reg.fit(x_train.reshape(-1,1),y_train)
```

```python
y_pred = svm_poly_reg.predict(x_train.reshape(-1,1))
# plot training data
plt.scatter(x_train,y_train,label="training data")
plt.scatter(x_train,y_pred,label="model")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
